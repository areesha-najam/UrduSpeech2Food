```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_excel("urdu-recipes.xlsx")
```

```python
import pandas as pd

# drop duplicates by "id" column and keep only the first occurrence
df = df.drop_duplicates(subset=["Id"], keep="first")

# count occurrences of "source" values
hummasala_count = df.loc[df["source"] == "hum masala", "source"].count()
k_foods_count = df.loc[df["source"] == "kfoods", "source"].count()

print(f"Number of entries with source = hummasala: {hummasala_count}")
print(f"Number of entries with source = k foods: {k_foods_count}")
```

```python
# extract the column with the ingredients
ingredients = df["ing_name"]

# use the unique() method to find the unique ingredients
unique_ingredients = ingredients.unique()

# print the unique ingredients
Ing=[]
print("Unique ingredients:")
for ingredient in unique_ingredients:
    print(ingredient)
    Ing.append(ingredient)
print("Total unique ingredients: ", len(Ing)-1)
```

```python
# extract the column with the ingredients
ingredients = df["Ing_unit"]

# use the unique() method to find the unique ingredients
unique_ingredients = ingredients.unique()

# print the unique ingredients
IngUnit=[]
print("Unique ingredients uniit:")
for unit in unique_ingredients:
    print(unit)
    IngUnit.append(unit)
print("Total unique ingredients: ", len(IngUnit)-1)
```

```python
# count occurrences of each ingredient
ingredient_counts = df["ing_name"].value_counts()

# create a bar chart of the results
ingredient_counts.plot(kind="bar", figsize=(12, 6),  color="grey")

# set the axis labels and title
plt.xlabel("Ingredient")
plt.ylabel("Occurrences")
plt.title("Occurrences of Ingredients in Dataset")

# show the plot
plt.show()
```
